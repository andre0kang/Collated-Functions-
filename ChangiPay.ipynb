```python
import io
import pandas as pd
from google.colab import files
import numpy as np
import xlsxwriter
```

```python
uploaded = files.upload()
```

```python
xlsx_file = io.BytesIO(uploaded.get('CP_Performance_wk_20220515.xlsx'))
```

```python
df1 = pd.read_excel(xlsx_file, 0)
df2 = pd.read_excel(xlsx_file, 1)
df3 = pd.read_excel(xlsx_file, 2)
df4 = pd.read_excel(xlsx_file, 3)
df5 = pd.read_excel(xlsx_file, 4)
df6 = pd.read_excel(xlsx_file, 5)
df7 = pd.read_excel(xlsx_file, 6)
```

```python
df1
```

```python
df_list = np.split(df1, df1[df1.isnull().all(1)].index) 

for df1 in df_list:
    print(df1, '\n')
```

```python
writer = pd.ExcelWriter('Excel_one.xlsx', engine='xlsxwriter')
for i in range(1, len(df_list) + 1):
    df_list[i - 1] = df_list[i - 1].dropna(how='all')
    df_list[i - 1].to_excel(writer, sheet_name='Sheet{}'.format(i), header=None, index=False)
    
# Save the excel file
writer.save()
```
